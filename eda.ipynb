```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

### 导入数据

```python
csv_list = os.listdir("data")
all_data_frames = []
for csv in csv_list:
    csv_dir = "data/" + csv
    single_data = pd.read_csv(csv_dir, header=0)  # 读取csv
    # single_data = single_data.dropna(how='any')
    all_data_frames.append(single_data)  # 组合
all_data = pd.concat(all_data_frames, axis=0, ignore_index=True)
all_data.to_csv('all_data.csv', encoding='utf_8_sig')  # 保存整合的csv文件
all_data
```

### 去除空值

```python
all_data.dropna(axis=0, inplace=True)
all_data
```

### 去除重复

```python
all_data.drop_duplicates(inplace=True)
all_data
```

### 删除异常值


```python
all_data.loc[:, '面积'] != '面积'
```

```python
all_data = all_data[all_data.loc[:, '面积'] != '面积']
```

### 特征工程

### 计算每平方米面积

#### 楼层 1/18 分解为 总楼层、所在楼层

```python
all_data.loc[:, '所在楼层'] = all_data['楼层'].apply(lambda x: int(x.split('/')[0]))
all_data.loc[:, '总楼层'] = all_data['楼层'].apply(
    lambda x: int(x.split('/')[1].split('层')[0]))
```

#### 提取地铁信息 “距房山线良乡大学城西站550米” 提取有几个地铁，距离多少米

```python
# 删除 字段中 地铁：
all_data.loc[:, '地铁'] = all_data['地铁'].apply(lambda x: x.replace('地铁：', ''))
```

```python
# 正则表达式提取字段中的**米
import re
all_data.loc[:, '地铁距离'] = all_data['地铁'].apply(lambda x: re.findall(
    '(\d+)米', x)[-1] if re.findall('(\d+)米', x) else -1).astype('int64')
```

```python
all_data.loc[:, '地铁数量'] = all_data['地铁'].apply(
    lambda x: len(re.findall('线', x)))
```

### 部分特征字符映射为数值

```python
all_data['地区1映射'] = pd.factorize(all_data['位置1'])[0]
```

### 数据类型转换

```python
all_data = all_data.astype({'价格': 'int64', '面积': 'int64'})
```

```python
all_data.dtypes
```

```python
all_data['每平价格'] = all_data['价格']/all_data['面积']
all_data['每平价格'].astype('int')
```

#### 数据总体分析

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)  # 小数点后面保留1位小数
all_data.describe(include='all')
```

### 数据相关性分析

```python
num_data_list = ['价格', '每平价格','面积', '所在楼层', '总楼层', '地铁距离', '地铁数量', '地区1映射']
data = all_data[num_data_list]
```

```python
# 特征相关性，返回上三角矩阵
cormatrix = data.corr()
cormatrix *= np.tri(*cormatrix.values.shape, k=-1).T
```

```python
# 特征相关性重新堆叠stack
cormatrix = cormatrix.stack()
cormatrix
```

```python
# 特征相关性按大小重新排序
cormatrix = cormatrix.reindex(cormatrix.abs().sort_values(ascending=False).index).reset_index()
cormatrix
```

```python
import seaborn as sns

# Compute the correlation matrix 
corr_all = data.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr_all, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize = (11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_all, mask = mask,
            square = True, linewidths = .5, ax = ax, cmap = "BuPu")     

plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus']=False   #解决负号“-”显示为方块的问题 

plt.show()
```

### 区域分析

```python
# 各区域数量统计
all_data['位置1'].value_counts()
```

```python
# 统计个区的平均价格和可租房数量
price_location = all_data.groupby("位置1").agg({'价格':'mean','位置1':'count','每平价格':'mean'})
price_location
```

```python
# 修改列名
# 多层列名
# price_location.columns = [i[0] + "_" + i[1] for i in price_location.columns]
price_location.columns = ['价格平均','数量合计','每平价格']
price_location
```

```python
# 排序
price_sort = price_location.sort_values('价格平均',ascending=False)
count_sort = price_location.sort_values('数量合计',ascending=False)
unit_sort = price_location.sort_values('每平价格',ascending=False)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(28,5))
plt.subplot(1,3,1)
plt.bar(price_sort.index,price_sort.iloc[:,0])
plt.title('各区域的平均价格统计')

plt.subplot(1,3,2)
plt.bar(count_sort.index,count_sort.iloc[:,1])
plt.title('各区域租房数量统计')

plt.subplot(1,3,3)
plt.bar(unit_sort.index,unit_sort.iloc[:,2])
plt.title('各区域租房单价统计')
plt.show()
```

## 小区分析

### 没有呈现出相关规律，可以把连续数据离散化

```python
metro_num = all_data.groupby('地铁数量').agg({'每平价格':'mean'})
metro_num
```

```python
# 切分，使用cat函数
bins = [0,500,1000,1500,2000,10000]
cast = pd.cut(all_data['地铁距离'],bins,labels=['小于500米','500-1000米','1000-1500米','1500-2000米','2000米以上'])
cast.name = '地铁距离离散'
cast
```

```python
# 拼接
data1 = pd.concat([all_data,cast],axis=1)
data1 = data1.dropna()
data1
```

```python
metro_distance = data1.groupby('地铁距离离散').agg({'每平价格':'mean'})
metro_distance.index = metro_distance.index.astype(dtype='str')
metro_distance
```

```python
plt.figure(figsize=(24,8))
plt.subplot(1,2,1)
plt.scatter(metro_distance.index,metro_distance['每平价格'])
plt.plot(metro_distance.index,metro_distance['每平价格'])
plt.title('地铁的距离远近和价格之间关系图')

plt.subplot(1,2,2)
plt.scatter(metro_num.index,metro_num['每平价格'])
plt.plot(metro_num.index,metro_num['每平价格'])
plt.title('地铁的数量和价格之间关系图')
```
