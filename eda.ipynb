```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

### 导入数据

```python
csv_list = os.listdir("data")
all_data_frames = []
for csv in csv_list:
    csv_dir = "data/" + csv
    single_data = pd.read_csv(csv_dir, header=0)  # 读取csv
    # single_data = single_data.dropna(how='any')
    all_data_frames.append(single_data)  # 组合
all_data = pd.concat(all_data_frames, axis=0, ignore_index=True)
all_data.to_csv('all_data.csv', encoding='utf_8_sig')  # 保存整合的csv文件
all_data
```

### 去除空值

```python
all_data.dropna(axis=0, inplace=True)
all_data
```

### 去除重复

```python
all_data.drop_duplicates(inplace=True)
all_data
```

### 删除异常值


```python
all_data.loc[:,'面积']!='面积'
```

```python
all_data = all_data[all_data.loc[:,'面积']!='面积']
```

### 特征工程

#### 楼层 1/18 分解为 总楼层、所在楼层

```python
all_data.loc[:,'所在楼层'] = all_data['楼层'].apply(lambda x:int(x.split('/')[0]))
all_data.loc[:,'总楼层'] = all_data['楼层'].apply(lambda x:int(x.split('/')[1].split('层')[0]))
```

#### 提取地铁信息 “距房山线良乡大学城西站550米” 提取有几个地铁，距离多少米

```python
# 删除 字段中 地铁：
all_data.loc[:,'地铁'] = all_data['地铁'].apply(lambda x:x.replace('地铁：',''))
```
